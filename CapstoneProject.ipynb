```python
csv_file = r"D:\capstoneproject\craigslist_vehicles.csv" 
data = pd.read_csv(csv_file, delimiter=',')
data.head(2)
```

```python
conn = mysql.connector.connect(**db_config)
cursor = conn.cursor()
```

```python
table_name = 'craiglist_vehicles'
max_description_length = 1000  # Adjust according to your needs

# Create the table with appropriate column lengths
columns = ', '.join([f'`{col}` VARCHAR({max_description_length})' for col in data.columns])

create_table_query = f"""
CREATE TABLE IF NOT EXISTS {table_name} (
    {columns}
);
"""

cursor.execute(create_table_query)
```

```python
data = data.dropna(axis=1, how='all')
data = data.applymap(lambda x: str(x) if pd.notna(x) else None)

# Insert data into the table using parameterized queries
for _, row in data.iterrows():
    # Truncate the description if it exceeds the maximum length
    description = str(row.get('description', ''))[:max_description_length]

    # Update the 'description' column in the row
    row['description'] = description

    columns = ', '.join([f'`{col}`' for col in data.columns])
    values = ', '.join(['%s' for _ in range(len(data.columns))])
    
    insert_query = f"""
    INSERT INTO {table_name} ({columns})
    VALUES ({values});
    """
    
    # Execute the query with the row values as a tuple
    cursor.execute(insert_query, tuple(row))
```

```python
select_query = f"SELECT * FROM {table_name};"
cursor.execute(select_query)

# Fetch all the rows
result = cursor.fetchall()

# Display the results
for row in result:
    print(row)
```

```python
conn.commit()
cursor.close()
conn.close()
```
